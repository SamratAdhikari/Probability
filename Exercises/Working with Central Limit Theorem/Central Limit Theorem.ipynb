# <center>Central Limit Theorem</center>

### Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Central Limit Theorem in practice

```python
# Generate number of samples
num = [1, 10, 50, 100]
means = [] # sample means

# Generate 1, 10, 30, 100 random numbers from -50 to +50
for i in num:
    np.random.seed(0)
    x = [np.mean(np.random.randint(-50, 50, i)) for j in range(1000)]
    means.append(x)

k = 0

fig, ax = plt.subplots(2, 2, figsize=(9, 9))
for i in range(0, 2):
    for j in range(0, 2):
        sns.distplot(a=means[k], bins=10, kde=True, color='purple', ax=ax[i, j])
#         ax[i, j].distplot(means[k], bins=10, density=True)
        ax[i, j].set_title(label=f"Sample Size: {num[k]}")
        k += 1
plt.show()
```

Clearly, as we increase the sample size, the data tends to follow Normal Distribution.
