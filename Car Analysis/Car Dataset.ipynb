#### Que #1: If i choose a car randomly, what is the probability that it is a Sedan?

```python
df['carbody'].value_counts()
```

```python
E = df['carbody'].value_counts()['sedan']
S = df['carbody'].value_counts().sum()
P = round(E/S, 4)
print("The probability that I get a Sedan when choosen randomly is", P)
```

#### Que #2: What is the probability that the car choosen randomly is Sedan and has fuel type as Gas(petrol)?

```python
crosstab = pd.crosstab(df['carbody'], df['fueltype'], margins=True)
crosstab
```

```python
E = crosstab.loc['sedan', 'gas']
S = len(df)
P = round(E/S, 4)
print("The probability that I get a Sedan which uses Fuel when choosen randomly is", P)
```

#### Que #3: What is the probability that a car is Sedan, given it is a fuel type?

```python
crosstab = pd.crosstab(df['carbody'], df['fueltype'], margins=True)
crosstab
```

```python
# P(S|F) = P(S n F) / P(F)
pSnF = crosstab.loc['sedan', 'gas']
pF = len(df[df['fueltype'] == 'gas'])
pS_F = round(pSnF / pF, 4)

print("The probability that I get a Sedan given it uses Gas is", pS_F)
```
