# <center>Generate Discrete & Continuous Random Numbers</center>

### Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, binom, geom, uniform, expon, norm
import warnings
warnings.filterwarnings('ignore')
```

#### Generate discrete random numbers following Bernoulli Distribution

```python
data = bernoulli.rvs(size=10000, p=0.4)

sns.distplot(data)
plt.xlabel("Bernoulli")
plt.ylabel("Frequency")
plt.show()
```

Here, p = 0.4  
As (1-p): 60% of data take the value 0 (failure).  
As p: 40% of data take value 1 (success).

#### Generate discrete random numbers following Binomial Distribution

```python
data = binom.rvs(n=10, p=0.5, size=10000)

sns.distplot(data)
plt.xlabel("Binomial")
plt.ylabel("Frequency")
plt.show()
```

#### Generate discrete random numbers following Geometric Distribution

```python
data = geom.rvs(p=0.5, size=10000)

sns.distplot(data)
plt.xlabel("Geometric")
plt.ylabel("Frequency")
plt.show()
```

#### Generate continuous random numbers following Uniform Distribution

```python
data = uniform.rvs(loc=0, scale=10, size=10000)

sns.distplot(data, bins=100)
plt.xlabel("Uniform")
plt.ylabel("Frequency")
plt.show()
```

#### Generate continuous random numbers following Exponential Distribution

```python
data = expon.rvs(loc=0, scale=1, size=10000)

sns.distplot(data, bins=100)
plt.xlabel("Exponential")
plt.ylabel("Frequency")
plt.show()
```

#### Generate continuous random numbers following Normal Distribution

```python
data = norm.rvs(loc=0, scale=1, size=10000)

sns.distplot(data, bins=100)
plt.xlabel("Normal")
plt.ylabel("Frequency")
plt.show()
```
